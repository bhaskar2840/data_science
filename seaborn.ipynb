to visualize the statistical relationship we use 

-> relplot()  # this takes two axes relationship 

```python
df=sns.load_dataset('tips')
df.head()
```

```python
sns.relplot(x='total_bill',y='tip',data=df,kind='line')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df,kind='scatter')
plt.show() 

# by default only scatterplot is shown
```

```python
sns.relplot(x='total_bill',y='tip',data=df,hue='time')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df,hue='time',style='smoker')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df,hue='time',col='smoker')
plt.show()
```

```python
#we can use sns.scatterplot instead but the parameters are diff read the documentation.
```

```python
sns.scatterplot(x='total_bill',y='tip',data=df)
plt.show()
```

# category relation

like we have seen in relplot() we have numerical data that we were comparing.

but in case of categorical datasets were one entry is numerical and other is a category like gender , size of t-shirts etc 

catplot()   is used

```python
df=sns.load_dataset('tips')
df.head()
```

```python
sns.catplot(x='day',y='total_bill',data=df)
plt.show()
```

```python
sns.catplot(x='day',y='total_bill',data=df,jitter=0.005)
plt.show() #by default jitter is 1. it controlls the deviation.
```

```python
sns.catplot(x='day',y='total_bill',data=df,jitter=0.75)
plt.show()
```

```python
sns.catplot(x='day',y='total_bill',data=df,kind='swarm')
plt.show()

# it will not let overlap the points ,but remeber it is used for small data sets,
```

```python
sns.catplot(x='day',y='total_bill',data=df,kind='swarm',hue='sex')
plt.show()
```

```python
sns.catplot(x='sex',y='total_bill',data=df,kind='box')
plt.show()

#the dot points are the outliers in our plot.

# the mid line is 50 percentile and the box top is 75percentile 
# and below is 25percentile
```

```python
sns.catplot(x='sex',y='total_bill',data=df,kind='boxen')
plt.show()
```

```python
sns.catplot(x='sex',y='total_bill',data=df,kind='box',hue='day')
plt.show()
```

```python
sns.catplot(x='sex',y='total_bill',data=df,kind='violin')
plt.show()
```
